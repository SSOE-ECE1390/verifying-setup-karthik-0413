* Load an (Grayscale) image  (‘Data/Apollo_11_Launch.jpg’)
* View image
* Crop image
* Edit brightness of image
* Rotate image
* Save modified image

```python
import cv2
import matplotlib.pyplot as plt

# Load an (Grayscale) image ('Data/Apollo_11_Launch.jpg)
apollo_file = cv2.imread('Data/Apollo_11_Launch.jpg', cv2.IMREAD_GRAYSCALE)

# View Image
plt.imshow(apollo_file, cmap='gray')

# Crop Image
apollo_file = apollo_file[400:600, 600:800]

# Edit Brightness of Image
apollo_file = cv2.convertScaleAbs(apollo_file, alpha=1.5, beta=60)

# Rotate Image
apollo_file = cv2.rotate(apollo_file, cv2.ROTATE_180)

# Save Modified Image
plt.imshow(apollo_file, cmap='gray')
```
